```python
# bibliotecas
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi, sin, cos, tan, exp, sqrt
from scipy import optimize, integrate
import scipy.constants as cte
from scipy.constants import hbar
import functools
```

```python
# constantes
m = cte.m_e
a = 2e-9 # largura do poço
V0 = 1.25*cte.e # altura do poço
```

```python
k = lambda E: np.sqrt(2*m*E)/hbar
kappa = lambda E: np.sqrt(2*m*(V0-E))/hbar

epsilon = 1e-30
N = 10000 # n de pontos nos graficos
es = np.linspace(0 + epsilon, V0 - epsilon, N) # range da energia
```

```python
# funções do lado esquerdo e direito do caso par
Tangente = lambda E: tan(k(E)*a/2)
Raiz = lambda E: kappa(E)/k(E)

plt.plot(es/cte.e, Raiz(es), label=r"$\kappa/k$")
plt.plot(es/cte.e, Tangente(es), label=r"$\tan{(ka/2)}$")

plt.xlabel("E (eV)")
plt.axis([0, V0/cte.e, 0, 10])
plt.legend(loc = "upper center")
plt.grid(True)
plt.savefig('pot_fin_par.svg')
plt.show()
```

```python
# funções do lado esquerdo e direito do caso ímpar
Tangente = lambda E: tan(k(E)*a/2)
Raiz = lambda E: -k(E)/kappa(E)

plt.plot(es/cte.e, Raiz(es), label=r"$-k/\kappa$")
plt.plot(es/cte.e, Tangente(es), label=r"$\tan{(ka/2)}$")

plt.xlabel("E (eV)")
plt.axis([0, V0/cte.e, -10, 0])
plt.legend(loc = "lower center")
plt.grid(True)
plt.savefig('pot_fin_impar.svg')
plt.show()
```
